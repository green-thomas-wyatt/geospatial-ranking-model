```python
from pyspark.sql import SparkSession  # Before SC was the entry point, now it is Spark Session!!

# Constructs a session, runs locally using all available cores,
# names your Spark job, returns an existing session if one exists or creates a new one
spark = (
    SparkSession.builder
    .master("local[*]")
    .appName("Chapter 5: pyspark SQL")
    .getOrCreate()
)
```

```python
yelp_test = spark.read.json("data/Yelp JSON/yelp_dataset/yelp_academic_dataset_business.json")

yelp_test.show(5)

yelp_test.count()


```

```python
from pyspark.sql import functions as F


yelp_test.where(F.col("state") == "TX").show()
```

```python
yelp_test.filter(yelp_test.review_count > 25).show(5)
```

```python
yelp_test.orderBy("stars").show(5)
```

```python
yelp_test.select("name", "city", "state").show(5)
```
